```python
columns_to_fill = ['Credit Score', 'Annual Income']
train[columns_to_fill] = train[columns_to_fill].apply(pd.to_numeric, errors='coerce')
means = train[columns_to_fill].mean()
train[columns_to_fill] = train[columns_to_fill].fillna(means)
```

```python
train.isnull().sum()
```

```python
train.drop(columns='Years in current job',inplace=True)
```

```python
train
```

```python
train=train.dropna()
```

```python
train.isnull().sum()
```

```python
from sklearn.preprocessing import RobustScaler

columns_to_scale = ['Current Loan Amount', 'Annual Income', 'Monthly Debt', 'Current Credit Balance','Maximum Open Credit']  
scaler = RobustScaler()
train[columns_to_scale] = scaler.fit_transform(train[columns_to_scale])
```

```python
train
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
cat_columns=train.select_dtypes('object').columns
```

```python
for column in cat_columns:
    train[column] = encoder.fit_transform(train[column])
```

```python
train.columns
```

```python
train.columns = train.columns.str.replace(' ', '') 
```

```python
train
```

```python
X = train.drop('LoanStatus', axis=1)
y = train['LoanStatus']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(C=0.1,kernel='rbf')
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
model.score(X_test,y_test)
```
