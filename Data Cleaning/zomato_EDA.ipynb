# Zomato EDA 

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('zomato.csv')
df
```

```python
df.columns
```

```python
df=df.drop(columns=['url','address','dish_liked','reviews_list','menu_item','phone'])
df
```

```python
df.head(1)
```

```python
df.rename(columns={'listed_in(type)':'listed_type','listed_in(city)':'city','rest_type':'restaurant_type'},inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0,subset=['location','restaurant_type','cuisines','approx_cost(for two people)'])
df
```

```python
df['rate'].unique()
```

```python
def rate_clean(x):
    if (x=='NEW' or x=='-'):
        return np.nan
    else:
        x=str(x).replace('/5','')
        return float(x)
df['rate']=df['rate'].apply(rate_clean)
df
```

```python
df['approx_cost(for two people)'].unique()
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(',','')
df
```

```python
column=df.select_dtypes(include='object').columns
```

```python
column
```

```python
df['restaurant_type'].value_counts().plot(kind='bar',figsize=(18,5))
```

```python
df['restaurant_type'].value_counts()
```

```python
restaurant_type_others=df['restaurant_type'].value_counts()<1000
restaurant_type_others
```

```python
rest_less_than_1000=df['restaurant_type'].value_counts()[restaurant_type_others]
rest_less_than_1000
```

```python
def rest_type(x):
    if(x in rest_less_than_1000):
        return 'other'
    else:
        return x
```

```python
df['restaurant_type']=df['restaurant_type'].apply(rest_type)
df['restaurant_type'].value_counts()
```

```python
df['location'].value_counts()
```

```python
location_less_1000=df['location'].value_counts()<1000
location_less_1000
```

```python
location_less_than_1000=df['location'].value_counts()[location_less_1000]
```

```python
def location_less(x):
    if (x in location_less_than_1000):
        return 'other'
    else:
        return x
df['location']=df['location'].apply(location_less)
df['location'].value_counts()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines_300=df['cuisines'].value_counts()<300
cuisines_300
```

```python
cuisines_less_than_300=df['cuisines'].value_counts()[cuisines_300]
cuisines_less_than_300
```

```python
def cuisine(x):
    if (x in cuisines_less_than_300):
        return 'other'
    else:
        return x
df['cuisines']=df['cuisines'].apply(cuisine)
df['cuisines'].value_counts()
```
