```python
numerical_columns=df.select_dtypes(include=['int64','float64']).columns
numerical_columns
```

```python
categorical_columns=df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
num_impute=SimpleImputer(strategy='median')
col_impute=SimpleImputer(strategy='most_frequent')
```

```python
df[numerical_columns]=num_impute.fit_transform(df[numerical_columns])
```

```python
df[categorical_columns]=col_impute.fit_transform(df[categorical_columns])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
education_encoder = LabelEncoder()
workclass_encoder = LabelEncoder()
marital_status_encoder = LabelEncoder()
occupation_encoder = LabelEncoder()
relationship_encoder = LabelEncoder()
income_encoder=LabelEncoder()
gender_encoder=LabelEncoder()


df['education'] = education_encoder.fit_transform(df['education'])
df['workclass'] = workclass_encoder.fit_transform(df['workclass'])
df['marital-status'] = marital_status_encoder.fit_transform(df['marital-status'])
df['occupation'] = occupation_encoder.fit_transform(df['occupation'])
df['relationship'] = relationship_encoder.fit_transform(df['relationship'])
df['income']=income_encoder.fit_transform(df['income'])
df['gender']=gender_encoder.fit_transform(df['gender'])
```

```python
df
```

```python
df=df.drop(columns=['fnlwgt','race','capital-gain','capital-loss','native-country'])
df
```

# Model Building 

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns='income')
y=df['income']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

```python
classification=classification_report(y_test,y_pred)
print(classification)
```

```python
confusion=confusion_matrix(y_test,y_pred)
confusion
```
