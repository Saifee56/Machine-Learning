# Model Building

```python
X=df.drop(['TenYearCHD'],axis=1)
y=df['TenYearCHD']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit_transform(X)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
adaboost=AdaBoostClassifier(n_estimators=40,learning_rate=1,algorithm='SAMME',random_state=0)
```

```python
adaboost.fit(X_train,y_train)
```

```python
y_pred=adaboost.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy :", accuracy_score(y_test,y_pred))
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
param_dist={
    'n_estimators':randint(1,100),
    'learning_rate':[0.1,0.5,1.0,1.5,2.0],
    'algorithm':['SAMME','SAMME.R'],
    'random_state':[0]
}
```

```python
adaboost1=AdaBoostClassifier(random_state=0)
```

```python
random_search=RandomizedSearchCV(adaboost1,param_distributions=param_dist,n_iter=10,cv=5,n_jobs=1,random_state=0)
```

```python
random_search.fit(X,y)
```

```python
best_params=random_search.best_params_
best_estimator=random_search.best_estimator_
```

```python
best_params
```

```python
best_estimator
```

```python
print("Randomized Search Results:")
for i, (params, score) in enumerate (zip(random_search.cv_results_['params'], random_search.cv_results_['mean_test_score'])):
    print(f"Combination {i+1} - Parameters: {params}, Mean Test Score (Accuracy): {score:.4f}")
```

```python
accuracy = best_estimator.score(X_test, y_test)
```
