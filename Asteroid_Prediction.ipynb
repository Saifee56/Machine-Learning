```python
df_new=df.drop(columns=['id','name','pdes','prefix','equinox'],axis='columns',inplace=False)
```

```python
df_new.shape
```

```python
df_new = df_new.drop(columns = ['epoch_mjd', 'epoch_cal', 'ad', 'moid', 'moid_ld', 'tp_cal', 'per_y', 'sigma_ad', 'sigma_tp', 'sigma_ma','diameter','albedo','diameter_sigma','orbit_id'])
```

```python
df_new.isnull().sum()
```

```python
columns_to_fill=['H','sigma_e','sigma_a','sigma_q','sigma_i','sigma_om','sigma_w','sigma_n','sigma_per']
```

```python
df_new[columns_to_fill] = df_new[columns_to_fill].fillna(df_new[columns_to_fill].mean()) 
```

```python
df_new.isnull().sum()
```

```python
df_new=df_new.dropna()
```

```python
df_new.isnull().sum()
```

```python
df_new
```

```python
df_new['neo'].unique()
```

```python
df_new['class'].unique()
```

```python
df_encoded = pd.get_dummies(df_new[['class']], prefix=['class'])
df_new = pd.concat([df_new, df_encoded], axis=1)
```

```python
df_new=df_new.drop(columns='class')
```

```python
df_new.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
columns_label=['neo','pha']
```

```python
df_new[columns_label] = df_new[columns_label].apply(lambda col: encoder.fit_transform(col))
```

```python
df_new
```

# Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_new.drop(columns=['spkid', 'full_name', 'pha'])
y=df_new['pha']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
df_new.pha.value_counts()/df_new.shape[0]
```

# SMOTE : For balancing the dataset

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state= 42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train.ravel())
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(max_iter = 1000, penalty = 'l2', C = 1.0, solver = 'lbfgs')
```

```python
LR.fit(X_train_res,y_train_res)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
y_pred=LR.predict(X_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
print(accuracy)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF=RandomForestClassifier(n_estimators = 100, max_depth = 4, min_samples_split = 10)
```

```python
RF.fit(X_train_res,y_train_res)
```

```python
y_pred1=RF.predict(X_test)
```

```python
accuracy1=accuracy_score(y_test,y_pred1)
```

```python
print(accuracy1)
```
